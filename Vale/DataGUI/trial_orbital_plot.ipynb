```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create two sine wave
fs = 200
w1 = 10
w2 = 10
t = np.linspace(0, 1, 1*fs+1)

x = np.sin(2*np.pi*w1*t)
y = np.cos(2*np.pi*w2*t)

plt.figure()
plt.subplot(211)
plt.plot(t, x)
plt.subplot(212)
plt.plot(t, y)
```

```python
# plot combination of x and y plot
plt.figure()
plt.axes().set_aspect('equal')
plt.plot(x, y)
```
