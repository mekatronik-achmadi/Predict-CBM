```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from fitter import Fitter

# Assuming 'data' is your DataFrame containing vibration data and 'choices' is your list of power classes

# Step 1: Fit distributions to vibration data columns for each power class
vibration_data = data[['Governor Actual Speed', 'UGB X-Displacement', 'UGB Y-Displacement', 
                       'LGB X-Displacement', 'LGB Y-Displacement', 'TGB X-Displacement', 
                       'TGB Y-Displacement']].dropna()

# Dictionary to store fitted distributions for each power class
fitted_distributions = {}

for power_class in choices:
    # Fit distributions for the current power class
    fitter = Fitter(vibration_data[data['Power Class'] == power_class])
    fitter.fit()

    # Get the best fitting distribution
    best_fit_name = fitter.get_best(method='sumsquare_error')[0]
    fitted_distributions[power_class] = best_fit_name

# Step 2: Plot PDFs of the best fitting distributions for each power class
fig, axes = plt.subplots(nrows=len(vibration_data.columns), ncols=1, figsize=(10, 10))

for i, col in enumerate(vibration_data.columns):
    for power_class in choices:
        best_fit_name = fitted_distributions[power_class]
        best_dist = getattr(fitter, best_fit_name)
        sns.kdeplot(data=vibration_data[data['Power Class'] == power_class][col], ax=axes[i], label=f'Power Class: {power_class}, Best Fit: {best_fit_name}', shade=True)

    axes[i].set_title(f'Distribution of {col} Across Power Classes')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Density')
    axes[i].legend(title='Power Class')
    axes[i].grid()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a date range and some sample data
date_rng = pd.date_range(start='1/1/2010', end='31/12/2016', freq='T')
np.random.seed(42)
power_data = np.random.uniform(low=0, high=100, size=(len(date_rng),))
vibration_data = np.random.uniform(low=0, high=10, size=(len(date_rng), 5))  # 5 columns of vibration data
# Introduce some NaN values randomly
nan_indices = np.random.choice(len(date_rng), size=5000, replace=False)
power_data[nan_indices] = np.nan
vibration_data[nan_indices, :] = np.nan
df = pd.DataFrame(date_rng, columns=['timestamp'])
df['power'] = power_data
df[['vibration1', 'vibration2', 'vibration3', 'vibration4', 'vibration5']] = vibration_data

# Ensure the timestamp column is in datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Filter out NaN values in the 'power' column and vibration columns
df = df.dropna(subset=['power'] + list(df.columns[df.columns.str.startswith('vibration')]))

# Categorize power data into 3 classes based on its values
conditions = [
    (df['power'] < 33),
    (df['power'] >= 33) & (df['power'] < 66),
    (df['power'] >= 66)
]
choices = ['Low', 'Medium', 'High']
df['power_class'] = np.select(conditions, choices, default='N/A')

# Verify the categorization
print(df['power_class'].value_counts())

# Extract year from the timestamp
df['year'] = df['timestamp'].dt.year

# Calculate duration (in hours) for each power class per year
duration_per_class_per_year = df.groupby(['year', 'power_class']).size().unstack(fill_value=0) / 60

# Plotting
ax = duration_per_class_per_year.plot(kind='bar', stacked=True, figsize=(10, 7), color=['blue', 'orange', 'green'])

# Customizing the plot
plt.title('Duration of Power Classes Per Year')
plt.xlabel('Year')
plt.ylabel('Total Duration (hours)')
plt.legend(title='Power Class')

# Adding values on each stacked bar
for container in ax.containers:
    labels = [f'{v.get_height():.1f}' if v.get_height() > 0 else '' for v in container]
    ax.bar_label(container, labels=labels, label_type='center')

plt.show()

# Step 1: Calculate Statistical Parameters for Vibration Data
vibration_stats = df.groupby('power_class')[['vibration1', 'vibration2', 'vibration3', 'vibration4', 'vibration5']].agg(['mean', 'var', 'std']).reset_index()
print(vibration_stats)

# Step 2: Plot PDF of Vibration Data Columns by Power Class
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(20, 5), sharey=True)

for i, col in enumerate(['vibration1', 'vibration2', 'vibration3', 'vibration4', 'vibration5']):
    for cls in choices:
        sns.kdeplot(data=df[df['power_class'] == cls][col], ax=axes[i], label=f'Power Class: {cls}', shade=True)
    axes[i].set_title(f'Distribution of {col}\nAcross Power Classes')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Density')

plt.tight_layout()
plt.show()
```
